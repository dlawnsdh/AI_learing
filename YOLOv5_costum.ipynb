```python
%cd /
from glob import glob

train_img_list = glob('/content/exex/YOLOv5_ex.v1i.yolov5pytorch/train/images/*.jpg')
print(len(train_img_list))

val_img_list = glob('/content/exex/YOLOv5_ex.v1i.yolov5pytorch/valid/images/*.jpg')
print(len(val_img_list))
```

```python
with open('/content/exex/YOLOv5_ex.v1i.yolov5pytorch/train.txt', 'w') as f:
    f.write('\n'.join(train_img_list) + '\n')

with open('/content/exex/YOLOv5_ex.v1i.yolov5pytorch/val.txt', 'w') as f:
    f.write('\n'.join(val_img_list) + '\n')   
```

```python
import yaml

with open('/content/exex/YOLOv5_ex.v1i.yolov5pytorch/data.yaml', 'r') as f:
    data = yaml.safe_load(f)
print(data)

data['train'] = '/content/exex/YOLOv5_ex.v1i.yolov5pytorch/train.txt'
data['val'] = '/content/exex/YOLOv5_ex.v1i.yolov5pytorch/val.txt'

with open('/content/exex/YOLOv5_ex.v1i.yolov5pytorch/data.yaml', 'w') as f:
    yaml.dump(data, f)
print(data)
```
